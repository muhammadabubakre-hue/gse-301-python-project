```python
print("Part 4: Interactive Menu System")
students = [
    {"name": "Yusuf", "cgpa":4.5, "active": True, "outstanding":0},
    {"name": "Andrew", "cgpa":3.5, "active": True, "outstanding":0},
    {"name": "Sam", "cgpa":4.8, "active": False, "outstanding":0},
    {"name": "Mat", "cgpa":3.8, "active": True, "outstanding":1},
    {"name": "Mercy", "cgpa":4.0, "active": True, "outstanding":0},
    {"name": "Yusuf", "cgpa":1.5, "active": True, "outstanding":2}
]

while True:
  print("1. View all students")
  print("2. Add new student")
  print("3. Check eligibility for graduation")
  print("4. Find top performer")
  print("5. Exit")

  choice = input("Please select an option:")
  match choice:
    case "1":
      print("Viewing all students...")
      if not students:
        print("This student record does not exist")
  else:
    for index, student in enumerate(students, start=1):
      print(f"{index}. {student['name']}, CGPA: {student['cgpa']}, Active: {student['active']}")



    case "2":
      print("Adding new students...")
      name = input("Enter the students name")
      try:
        cgpa = float(input("Please enter the CGPA:"))
        outstanding = int(input("Please enter the number of outstanding courses:"))
        vactive = input("Is said student active? yes or no: ").lower()
        if active == "yes":
           active = True
        else:
           vactive = False
           new_stud = {
               "name": name,
               "cgpa": cgpa,
               "active": vactive,
               "outstanding": outstanding,
           }
           students.append(new_student)
           print("The student record has been added successfully")
      except: ValueError
           print("There was an invalid input, student record has not been added.")
    case "3":
      print("Checking the eligibility for graduation...")
    case "4":
      print("Finding top performer...")
    case "5":
      print("Exiting the system...")
      break
    case "6":
      print("Invalid choice, try again")

def eligibility(cgpa, outstanding, is_active):
  if cgpa >= 2.5 and outstanding == 0 and is_active:
    return True, "This student is currently eligible for graduation"
  else:
    return False, "This student is currently not eligible for graduation"
```
